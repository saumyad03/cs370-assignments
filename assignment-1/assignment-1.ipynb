```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

## Problem 1: Simulation (20 points)

### Problem 1A (15 points)

```python
# setting distribution and sampling parameters
means = [0, 2]
covariance_matrix = [[0.3, -1], [-1, 5]]
m = 200
# sampling
samples = np.random.multivariate_normal(means, covariance_matrix, m)
```

### Problem 1B (5 points)

```python
# creating points for contour plot
x1, x2 = np.meshgrid(np.linspace(-2, 2, 100), np.linspace(-5, 9, 100))
positions = np.dstack((x1, x2))
# creating prob. distribution and evaluating pdf values
distribution = multivariate_normal(means, covariance_matrix)
pdf = distribution.pdf(positions)
# plotting contour plot
plt.contour(x1, x2, pdf)
# superimposing scatterplot on top of contour plot
plt.scatter(samples[:, 0], samples[:, 1], color="purple")
# labels
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Problem 1B Plot")
plt.show()
```

## Problem 2: Projection (20 points)

### Problem 2A (5 points)

The principal components can be determined by performing singular value decomposition on the covariance matrix. After doing so, the principal components are the columns of the V matrix (transpose of third matrix in SVD). Ordered by the magnitudes of their corresponding singular values (which are prsent in the second matrix in SVD), the principal components represent the directions of maximum variance in the data. The principal components are also the eigenvectors of the covariance matrix. To determine, the first 2 principal components, we will use Python:

```python
covariance_matrix = np.array([[4, 2, 1],
                              [2, 3, 1.5],
                              [1, 1.5, 2]])
# performs singular value decomposition
U, S, V_t = np.linalg.svd(covariance_matrix)
# extract first two principal components from V matrix
V = V_t.T
first_component = V[:, 0]
second_component = V[:, 1]
print(first_component)
print(second_component)
```

### Problem 2B (5 points)

First, we must simulate a 3D Gaussian random vector with the aforementioned covariance matrix by sampping 1000 3D vectors from the distribution. I will assume the mean vector is full of 0's.

```python
mean_vector = np.zeros(3)
m = 1000
samples = np.random.multivariate_normal(mean_vector, covariance_matrix, size=m)
```

In order to project the simulated vectors onto the subspace represented by the first 2 principal components, we must take the dot product of each simulated vector with each principal component. This captures the component of the vector that points in the direction of the principal component.

```python
# generate projected samples
projection_matrix = np.column_stack((first_component, second_component))
projected_samples = np.dot(samples, projection_matrix)
# plot projected samples
plt.scatter(projected_samples[:, 0], projected_samples[:, 1], alpha=0.5)
plt.title("2D Projection of 3D Samples")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.show()
```

### Problem 2C (10 points)

In order to reverse the projection to map back to the original 3D space, we need to apply the transformation that is opposite to the one represented by the projection matrix. We can do this by simply multiplying the projected samples by the inverse of the projection matrix. This will map the samples from the 2D subspace back tot he original 3D spcae. Then, we will visualize the reconstructed points with a scatterplot.

```python
inverse_projection_matrix = np.linalg.pinv(projection_matrix)
reconstructed_samples = np.dot(projected_samples, inverse_projection_matrix)

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=reconstructed_samples[:, 0],
    y=reconstructed_samples[:, 1],
    z=reconstructed_samples[:, 2],
    mode='markers',
    #marker=dict(size=6, opacity=0.5, color='blue'),
    #name='Reconstructed Samples'
))
fig.show()
```

## Problem 3: Stochastic Gradient Descent (30 points)

Since we were expected to use the linear regression example from the class notes, I copied the implementations for generating the dataset from [the course site](https://pantelis.github.io/artificial-intelligence/aiml-common/lectures/regression/linear-regression/linear_regression.html). I modified the example so that I had 100 examples for training instead of just 10.

```python
def create_toy_data(func, sample_size, std, domain=[0, 1]):
    x = np.linspace(domain[0], domain[1], sample_size)
    np.random.shuffle(x)
    y = func(x) + np.random.normal(scale=std, size=x.shape)
    return x, y

def sinusoidal(x):
    return np.sin(2 * np.pi * x)

x_train, y_train = create_toy_data(sinusoidal, 100, 0.25)
x_test = np.linspace(0, 1, 100)
y_test = sinusoidal(x_test)
```

Then, I implemented Polynomial Regression along with a from-scratch Stochastic Gradient Descent. The following were my hyperparameter values:
- degree: 10
- epoch number: 1000 (note that loops iterate number of epochs * number of examples times as one epoch is an entire iteration through the training set)
- learning rate: 0.1
These choices were largely based off trial and error for what generalized the best.

```python
m = len(x_train) # number of examples
# first, we create polynomial features w/ training/test dataset
degree = 10
poly_x_train = np.vander(x_train, degree + 1, increasing=True)
poly_x_test = np.vander(x_test, degree + 1, increasing=True)
# then, we set our hyperparameters
num_epochs = 1000 # passes over entire training set
eta = 0.1 # learning rate
# then, we set our parameters
w = np.random.randn(degree + 1, 1)
# to make predictions, we take the dot product of
# our parameters and polynomial features
def predict(w, x):
    return np.dot(w.T, x.T)
# to calculate our loss, we use mean squared error
def loss(w, x, y, m):
    return np.sum(np.square(predict(w, x) - y)) / m
# variables to track info for plotting loss vs. epoch
epochs = [i + 1 for i in range(num_epochs)]
losses = []
# loop using loss function and optimization algorithm
# to improve parameters
for epoch in range(num_epochs):
    for iteration in range(m):
        # stochastic -> randomly select 1 example
        random_index = np.random.randint(m)
        xi = poly_x_train[random_index : random_index + 1]
        yi = y_train[random_index : random_index + 1]
        # calculate gradient
        prediction = predict(w, xi)
        diff = prediction - yi
        gradient = np.array([(diff[0][0] * poly_feature)/m for poly_feature in xi]).T
        # tweak parameters
        w = w - eta * gradient
    # tracks losses for plotting
    losses.append(loss(w, poly_x_train, y_train, m))
# plots loss vs. epoch for training and test dataset
plt.plot(epochs, losses)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, "-k", label="$\sin(2\pi x)$")
predictions_b = predict(w, poly_x_test).T
plt.plot(x_test, predictions_b, "-r", label="Baseline SGD")
plt.legend()
plt.show()
```

## Problem 4: SGD Enhancements (30 points)

First, we will implement Momentum. In order to understand how to implement this algorithm, I used [this video](https://www.youtube.com/watch?v=NE88eqLngkg&ab_channel=DeepBean). This involves replacing the gradient computation with a leaky average. The instantaneous gradient is replaced by one averaged over multiple past gradients. This is done by changing the parameters by a velocity per time step. This velocity changes in accordance with a mometum hyperparameter as well as the learning rate. Momentum characterizes the resistance of the velocity to change while the learning rate characterizes the influence of the gradient. To implement this, we will modify the code from Problem 3. The following were my hyperparameter values:
- degree: 10
- epoch number: 1000
- learning rate: 0.1
- momentum: 0.9

As we can see in the plot produced below, SGD w/ Momentum converges much more quickly than the Baseline SGD algorithm.

```python
# then, we set our parameters
w_m = np.random.randn(degree + 1, 1)
# variables to track info for plotting loss vs. epoch
losses_m = []
# initialize state variables and hyperparameter
velocity = 0
momentum = 0.9
# loop using loss function and optimization algorithm
# to improve parameters
for epoch in range(num_epochs):
    for iteration in range(m):
        # stochastic -> randomly select 1 example
        random_index = np.random.randint(m)
        xi = poly_x_train[random_index : random_index + 1]
        yi = y_train[random_index : random_index + 1]
        # calculate gradient
        prediction = predict(w_m, xi)
        diff = prediction - yi
        gradient = np.array([(diff[0][0] * poly_feature)/m for poly_feature in xi]).T
        # tweak parameters
        velocity = momentum * velocity - eta * gradient
        w_m = w_m + velocity
    # tracks losses for plotting
    losses_m.append(loss(w_m, poly_x_train, y_train, m))
# plots loss vs. epoch for training and test dataset
plt.plot(epochs, losses_m, label="SGD w/ Momentum")
plt.plot(epochs, losses, label="Baseline SGD")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, "-k", label="$\sin(2\pi x)$")
predictions_m = predict(w_m, poly_x_test).T
plt.plot(x_test, predictions_m, "-r", label="Momentum")
plt.legend()
plt.show()
```

Next, we will implement Adam. I used [the same video](https://www.youtube.com/watch?v=NE88eqLngkg&ab_channel=DeepBean) along with the [Deep Learning Textbook](https://d2l.ai/chapter_optimization/adam.html) to understand how to implement this algorithm. This algorithm scales the gradient step to adapt the learning rate. This is done by using exponential weighted moving averages. Therefore, it requires two state variables which we'll call v and s that are parameterized by beta_one and beta_two (nonnegative weighting parameters). For these, I used the recommended values from the textbook. We also have to renormalize the terms to address the bias towards smaller values that initializing the state variables to 0 brings. Finally, we also have a hyperparameter epsilon that acts as a numerical stabilizer, preventing us from dividng by numbers very close to 0. This will also be set to the recommended value. To implement this algorithm, we will modify the code from Problem 3. The following were my hyperparameter values:
- degree: 10
- epoch number: 1000
- learning rate: 0.1
- beta_one: 0.9
- beta_two: 0.999
- epsilon: 0.000001

Similar to Momentum, SGD w/ Adam converges much more quickly than just Baseline SGD as we can see in the plot produced below.

```python
# then, we set our parameters
w_a = np.random.randn(degree + 1, 1)
# variables to track info for plotting loss vs. epoch
losses_a = []
# initialize state variables and hyperparameters
v = 0
s = 0
beta_one = 0.9
beta_two = 0.999
epsilon = 0.000001
# loop using loss function and optimization algorithm
# to improve parameters
for epoch in range(num_epochs):
    for iteration in range(m):
        # stochastic -> randomly select 1 example
        random_index = np.random.randint(m)
        xi = poly_x_train[random_index : random_index + 1]
        yi = y_train[random_index : random_index + 1]
        # calculate gradient
        prediction = predict(w_a, xi)
        diff = prediction - yi
        gradient = np.array([(diff[0][0] * poly_feature)/m for poly_feature in xi]).T
        # update state variables_a
        v = beta_one * v + (1 - beta_one)*gradient
        s = beta_two * s + (1 - beta_two)*np.square(gradient)
        # normalizing state variables
        v_hat = v / (1 - beta_one)
        s_hat = s / (1 - beta_two)
        # rescaling gradient
        g = (eta * v_hat) / (np.sqrt(s_hat) + epsilon)
        # tweak parameters
        w_a = w_a - g
    # tracks losses for plotting
    losses_a.append(loss(w_a, poly_x_train, y_train, m))
# plots loss vs. epoch for training and test dataset
plt.plot(epochs, losses_a, label="SGD w/ Adam")
plt.plot(epochs, losses, label="Baseline SGD")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, "-k", label="$\sin(2\pi x)$")
predictions_a = predict(w_a, poly_x_test).T
plt.plot(x_test, predictions_a, "-r", label="Adam")
plt.legend()
plt.show()
```
